```python
%matplotlib inline
import matplotlib.pyplot as plt


x = [1, 2, 3, 4]
y = [4, 3, 2, 1]
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, y, "-o")

fig.savefig("myfigure.png")
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True)

ax0.plot(x, y)
ax1.plot(y, x)

ax0.set_title("ax0")
ax1.set_title("ax1")
fig.suptitle("My fig")
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "polar"})

ax.plot(x, y);
```

```python
import cartopy.crs as ccrs


fig, ax = plt.subplots(
    figsize=(6, 12),
    subplot_kw={"projection": ccrs.PlateCarree()}
)

ax.coastlines(resolution="110m", color="crimson")
```

```python
fig, ax = plt.subplots(
    figsize=(6, 12),
    subplot_kw={"projection": ccrs.PlateCarree()}
)

ax.coastlines(resolution="10m", color="cornflowerblue")
ax.set_extent([-54, -49, -33, -29])
```
